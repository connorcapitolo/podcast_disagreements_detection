```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# loading in the pickle file
df = pd.read_pickle("../outputs/compiled_annotations_df.pkl")
print(df.head())
```

```python
display(df['annotator'].unique())
display(df['text'].unique())
display(df['text'].nunique())
# 0pI
# 7vx
# 7r3
# 7Lx
# 0Tk
# 1Xg
# 2hg
# 1jf
# 2DC
# 6ho
# 79a
```

```python
# extracting only the podcasts we've annotated
annotated_pods = df.loc[(df['text'] != '2KRFxcWULTtVe0LXBjXduT') & (
    df['text'] != '73qaKP1ZaRmlOwhAwBHrug'), :]
display(annotated_pods['text'].unique())
display(annotated_pods['text'].nunique())
```

```python
annotated_pods['disagree_length'] = annotated_pods['end'] - annotated_pods['start']
display(annotated_pods.head())
avg_disagree_length = np.average(annotated_pods['disagree_length'])
print(f'Average length of disagreement: {avg_disagree_length: .4f} seconds')
display(annotated_pods['disagree_length'].describe())
```

```python
# interesting that Morris typically had the longest disagremeent clips
annotated_pods.loc[annotated_pods['disagree_length'] > 3.35, :].sort_values('disagree_length', ascending = False)
```

```python
print(f'We have about 206 minutes of audio that contains disagreement. If we used just average time chunks, how much audio many observations would we approximately have? \nAnswer: {206 * 60 / avg_disagree_length:.0f}')
```
