## Visualize all annotated episodes

Prerequisite steps:
1. Annotate episodes in prodigy, using the instructions in audio_annotation/README.md
2. Run compile_annotations.py with appropriate args, to produce dataframe "compiled_annotation_df"

```python
import seaborn as sns
import sys, os
import pandas as pd
import matplotlib.pyplot as plt

# import common analysis functions we've defined
parent_directory = os.path.dirname(os.path.dirname(os.getcwd()))
sys.path.append(parent_directory)
from common.annotation_utils import tp, get_union, plot_annotations

# read compiled annotations dataframe
df = pd.read_pickle("../outputs/compiled_annotations_df.pkl")

# define colors for plotting, for each annotator of interest
annotator_colors = {
    'connor': 'orange',
    'jiahui': 'blue',
    'morris': 'red',
    'max': 'purple'
}

# produce one plot for each episode ID
episodes = df['text'].unique().copy()
print(f"Number of episodes annotated: {len(episodes)}")

for e in episodes:
    plot_annotations(df, e, annotator_colors)
```
